```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs # this tool allows us to randomly generate data
# this is useful when practicing algorithms
```

```python
data = make_blobs(n_samples = 200, n_features = 2, centers = 4, cluster_std = 1.8, random_state = 101)
```

```python
data # data is a tuple 
```

```python
plt.scatter(data[0][:,0], data[0][:,1], c = data[1], cmap = 'rainbow')
```

```python
data[1]
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 4) # we made this data so we know there are 4
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
fig , (ax1, ax2) = plt.subplots(1, 2, sharey = True, figsize = (10,6))
# creating a sub plot of 1 row by 2 columns that share the same axies
# (ax1, ax2) is using tuple unpaking to grab axies
ax1.set_title('K_Means')
ax1.scatter(data[0][:,0], data[0][:,1], c = kmeans.labels_, cmap = 'rainbow')

ax2.set_title('Origionals')
ax2.scatter(data[0][:,0], data[0][:,1], c = data[1], cmap = 'rainbow')
```
